## Import from last week's notebook

```python
# Import Libraries
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib import font_manager
```

```python
# Load the data
census_df = pd.read_csv('Data/US_Census_LA_County.csv', dtype={'FIPS':str})
without_vehicle_gdf = gpd.read_file('Data/Without_Vehicle.geojson')
```

```python
# Merge the datasets on the FIPS code
census_df2 = census_df [['FIPS', 'NAME']]
merged_gdf = without_vehicle_gdf.merge(census_df2, right_on='FIPS', left_on='tract', how='inner')
```

```python
# Filter out Channel Islands tracts
merged_gdf = merged_gdf[~merged_gdf['FIPS'].isin(['06037599100', '06037599000'])]
```

## Refined Data Visualizations

### Histogram refined with custom colors, fonts, lineweights, etc.

```python
# Set custom colors
bin_color = '#f89739'
text_color = 'black'

# Define font properties
font_path_fira_light = 'Fonts/FiraSansExtraCondensed-Light.ttf'  
font_path_fira_regular = 'Fonts/FiraSansExtraCondensed-Regular.ttf'
font_path_fira_medium = 'Fonts/FiraSansExtraCondensed-Medium.ttf' 

fira_light = font_manager.FontProperties(fname=font_path_fira_light, size=12)
fira_regular = font_manager.FontProperties(fname=font_path_fira_regular, size=14)
fira_medium = font_manager.FontProperties(fname=font_path_fira_medium, size=16)

# Histogram of the percentage of households without a vehicle
plt.figure(figsize=(10, 4))
plt.hist(merged_gdf['no_vehicle_pct'], bins=30, color=bin_color, edgecolor='black', alpha=0.7)
plt.title('Distribution of Households Without a Vehicle', color=text_color, fontproperties=fira_medium)
plt.xlabel('Percentage of Households Without a Vehicle', color=text_color, fontproperties=fira_regular)
plt.ylabel('Frequency', color=text_color, fontproperties=fira_regular)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xticks(fontsize=10, color=text_color, fontproperties=fira_light)
plt.yticks(fontsize=10, color=text_color, fontproperties=fira_light)
plt.show()

# can't seem to figure out why the title font size doesn't change with assigned font size
```

### Box plot refined with custom colors, fonts, lineweights, etc.

```python
# Set custom colors
box_color = '#f89739'
text_color = 'black'

# Define font properties
font_path_fira_light = 'Fonts/FiraSansExtraCondensed-Light.ttf'  
font_path_fira_regular = 'Fonts/FiraSansExtraCondensed-Regular.ttf'
font_path_fira_medium = 'Fonts/FiraSansExtraCondensed-Medium.ttf' 

fira_light = font_manager.FontProperties(fname=font_path_fira_light, size=12)
fira_regular = font_manager.FontProperties(fname=font_path_fira_regular, size=14)
fira_medium = font_manager.FontProperties(fname=font_path_fira_medium, size=16)

# Box plot
plt.figure(figsize=(10, 2))
plt.boxplot(merged_gdf['no_vehicle'], vert=False, patch_artist=True, boxprops=dict(facecolor=box_color))
plt.title('Box Plot of Households Without a Vehicle', color=text_color, fontproperties=fira_medium)
plt.xlabel('Households Without a Vehicle', color=text_color, fontproperties=fira_regular)
plt.xticks(color=text_color, fontproperties=fira_light)
plt.yticks(color=text_color, fontproperties=fira_light)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()

# can't seem to figure out why the title font size doesn't change with assigned font size
```
